```python
import pandas as pd
from spellchecker import SpellChecker
from datetime import datetime, timedelta
from unidecode import unidecode
import re


# đọc file csv, kiểu đọc này phải để file trong cùng thư mục với code, nếu không phải trích dẫn đường link vd:'C:\user\....\data_for_test_10000.csv'
df = pd.read_csv('data_for_test_10000.csv')

# xem thông tin của dữ liệu để xem các cột và dữ liệu có giá trị thiếu
print(df.info())

# chỗ này là đặt tùy chọn hiển thị cho pandas
pd.set_option('display.float_format', lambda x: '%.0f' % x)

#-----------------------------------------------------------------------------

def convert_scientific_to_integer(text):
    try:
        # convert từ float sang int
        number = float(text)
        return int(number)
    except ValueError:
        # nếu không thể chuyển đổi, trả lại giá trị ban đầu
        return text

# áp dụng hàm vào cột 'ID'
df['ID'] = df['ID'].apply(convert_scientific_to_integer)

#-----------------------------------------------------------------------------
# dictonary liệt kê những trường hợp đa số xuất hiện trong file, vì tên k có dấu
names = {'Nguyen': 'Nguyễn', 'Tran' : 'Trần', 'Van' : 'Văn', 'Ngan' : 'Ngân', 'Duc' : 'Đức', 'Lien' : 'Liên',
         'Thi' : 'Thị', 'Phuong' : 'Phương', 'Quoc' : 'Quốc', 'Bich' : 'Bích', 'Hue' : 'Huệ', 'Hanh' : 'Hạnh', 'Chau' : 'Châu',
         'Quynh' : 'Quỳnh', 'Viet' : 'Việt', 'Hieu' : 'Hiếu', 'Tuan' : 'Tuấn', 'Bui' : 'Bùi', 'Le' : 'Lê', 'Vu' : 'Vũ', 'Hoang' : 'Hoàng' 
        }

# hàm đổi tên
def replace_names(text, names_dict):

    text = re.sub(r'[!@#$%^&*()_+{}|:"<>?`~\-=\\\[\];,\./]', '', text)
    
    for key, value in names_dict.items():
        text = text.replace(key, value)
    return text

df['Fullname'] = df['Fullname'].apply(lambda x: replace_names(x, names))

#-----------------------------------------------------------------------------

def fix_date_format(date_str):
    # kiểm tra nếu là số tuổi, chuyển đổi thành ngày tháng năm sinh
    if re.match(r'^\d+$', date_str):
        age = int(date_str)
        birth_year = pd.Timestamp.now().year - age
        return f"01/01/{birth_year}"

    # kiểm tra và chuyển đổi các định dạng ngày tháng năm
    formats = [
        r'^(\d{4})(\d{2})(\d{2})$',  # 19900101
        r'^(\d{1,2})/(\d{1,2})/(\d{4})$',  # 1/1/1990 hoặc 01/01/1990
        r'^(\d{4})/(\d{2})/(\d{2})$',  # 1990/01/01
        r'^(\d{1,2})-(\d{1,2})-(\d{4})$',  # 15-07-1985
        r'^(\d{2})(\d{2})(\d{4})$',  # 12041990
        r'^(\d{2})/(\d{2})/(\d{4})$',  # 31/02/1991 hoặc 12/5/1990
    ]

    for pattern in formats:
        match = re.match(pattern, date_str)
        if match:
            day, month, year = match.groups()
            return f"{day.zfill(2)}/{month.zfill(2)}/{year}"

    # nếu không khớp với bất kỳ định dạng nào, trả về None
    return None
    
df['Birthday'] = df['Birthday'].apply(fix_date_format)

#-----------------------------------------------------------------------------

def title_sex(text):
    try:
        clean_text = text.lower().strip()
        
        return clean_text.title()

    except:
        return None

df['Sex'] = df['Sex'].astype(str).apply(title_sex)

#-----------------------------------------------------------------------------

# danh sách các tỉnh thành Việt Nam
provinces = [
    'hà nội', 'hồ chí minh', 'đà nẵng', 'hải phòng', 'cần thơ',
    'an giang', 'bà rịa - vũng tàu', 'bạc liêu', 'bắc kạn', 'bắc giang',
    'bắc ninh', 'bến tre', 'bình định', 'bình dương', 'bình phước',
    'bình thuận', 'cà mau', 'cao bằng', 'đắk lắk', 'đắk nông',
    'điện biên', 'đồng nai', 'đồng tháp', 'gia lai', 'hà giang',
    'hà nam', 'hà tĩnh', 'hải dương', 'hậu giang', 'hòa bình',
    'hưng yên', 'khánh hòa', 'kiên giang', 'kon tum', 'lai châu',
    'lâm đồng', 'lạng sơn', 'lào cai', 'long an', 'nam định',
    'nghệ an', 'ninh bình', 'ninh thuận', 'phú thọ', 'phú yên', 'quảng bình',
    'quảng nam', 'quảng ngãi', 'quảng ninh', 'quảng trị', 'sóc trăng',
    'sơn la', 'tây ninh', 'thái bình', 'thái nguyên', 'thanh hóa',
    'thừa thiên huế', 'tiền giang', 'trà vinh', 'tuyên quang',
    'vĩnh long', 'vĩnh phúc', 'yên bái'
]

# kiểm tra và sửa định dạng tỉnh thành
def clean_hometown(text):
    try:
        # loại bỏ dấu tiếng Việt và chuyển đổi thành chữ thường
        clean_text = unidecode(text).lower().strip()

        # xóa chữ "thành phố" nếu nó xuất hiện
        clean_text = clean_text.replace('thanh pho ', '')
        
       # kiểm tra và sửa định dạng
        for province in provinces:
            if clean_text == unidecode(province).lower():
                return province.title()

        return None
    
    except:
        return None

df['Hometown'] = df['Hometown'].astype(str).apply(clean_hometown)

#-----------------------------------------------------------------------------

# lương
df['Salary'] = df['Salary'].str.replace(r'[^0-9]', '', regex=True).astype(float)

#-----------------------------------------------------------------------------

# khởi tạo spellChecker
spell = SpellChecker()

# sửa chữa lỗi chính tả
def correct_spelling(text):
    corrected_text = []
    words = text.split()
    
    for word in words:
        corrected_word = spell.correction(word)
        corrected_text.append(corrected_word.title())
    
    return ' '.join(corrected_text)

df['Marital Status'] = df['Marital Status'].apply(lambda x: correct_spelling(str(x)))

# Xử lý các ký tự không mong muốn khác
df['Marital Status'] = df['Marital Status'].str.replace(r'[^a-zA-Z\s]', '', regex=True)

#------------------------------------------------------------------------------

# lưu lại file sau khi đã tiền xử lý
df.to_csv('processed_file.csv', index=False)

# kiểm tra lại thông tin của dữ liệu sau khi tiền xử lý
print(df.info())
```

```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

file_path = 'processed_file.csv'  
df = pd.read_csv(file_path)

# chuyển sang Datetime
df['Birthday'] = pd.to_datetime(df['Birthday'], format='%d/%m/%Y', errors='coerce')

# xóa data không chuyển được
df = df.dropna(subset=['Birthday'])

# tính tuổi từng người
df['Birthday'] = (pd.Timestamp('now').year - df['Birthday'].dt.year) - np.vectorize(lambda x, y: 1 if x < y else 0)(pd.Timestamp('now').month, df['Birthday'].dt.month) - np.vectorize(lambda x, y: 1 if x < y else 0)(pd.Timestamp('now').day, df['Birthday'].dt.day)

# tạo biểu đồ
plt.figure(figsize=(20, 15))
ax=sns.countplot(data=df, y='Birthday', order=df['Birthday'].value_counts().index)
plt.title('Phân bố tuổi')
plt.xlabel('Số lượng')
plt.ylabel('Tuổi')

# hiển thị dữ liệu trên từng cột
for p in ax.patches:
    ax.text(p.get_width(), p.get_y() + p.get_height() / 2., '%d' % int(p.get_width()), 
            fontsize=12, color='black', ha='left', va='center')
    
plt.show()
```
